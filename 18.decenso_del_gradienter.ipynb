```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
_X = np.linspace(-2,2,100)
_Y = np.linspace(-2,2,100)
_Z = np.zeros((100,100))

func = lambda th: np.sin(((th[0]**2)/2)+(((th[1]**2)/4)+3))*np.cos(2*th[0]+1*(np.e**th[1]))
# en el original es en cos(1-e**t[1]), per cambiaremos este datp
```

```python
for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = func((x,y))
```

```python
plt.contour(_X,_Y,_Z)
plt.show()
```

```python
plt.contour(_X,_Y,_Z,100)
plt.show()
```

```python
plt.contourf(_X,_Y,_Z)
plt.colorbar()

_I = np.random.rand(2) * 4 - 2

_DIF = 0.0001
_LR = 0.015
_C = np.copy(_I)
_GRAD = np.zeros(2)

plt.plot(_GRAD[0], _GRAD[1],'X',c='black')

for _ in range(10000):
    for it, th in enumerate(_I):
        
        _C = np.copy(_I)
        _C[it] = _C[it] + _DIF
        _NDER = (func(_C)-func(_I))/_DIF
        #print(_C[it], _NDER)
        _GRAD[it] = _NDER
        
    _I = _I - _LR*_GRAD
    
    if _ % 5 == 0:
        plt.plot(_I[0], _I[1], '.', c='blue')
        
plt.plot(_I[0], _I[1], 'o', c='white')
plt.show()
```
