```python
import numpy as np
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
_X, _Y = mglearn.datasets.load_extended_boston()

x_train, x_test, y_train, y_test = train_test_split(_X, _Y, random_state=0)
```

```python
from sklearn.linear_model import Ridge
```

```python

ridge = Ridge().fit(x_train, y_train)
print('{:.2f}, {:.2f}'.format(ridge.score(x_train, y_train), ridge.score(x_test, y_test)))
```

```python
lr = LinearRegression().fit(x_train,y_train)
lr.score(x_test, y_test)
```

```python

ridge01 = Ridge(alpha=.1).fit(x_train, y_train)
print('{:.2f}, {:.2f}'.format(ridge01.score(x_train, y_train), ridge01.score(x_test, y_test)))
```

```python
ridge10 = Ridge(alpha=10).fit(x_train, y_train)

print(ridge10.score(x_train,y_train),ridge10.score(x_test,y_test))
```

```python
import matplotlib.pyplot as plt

plt.plot(ridge.coef_,'s', label='ridge alpha 1')
plt.plot(ridge10.coef_, '^', label='ridge alpha 10')
plt.plot(ridge01.coef_, 'v', label='ridge alpha .1')

plt.plot(lr.coef_, '.',label='LinearRegression')
plt.xlabel('coefficients index')
plt.ylabel('coefficients magnitude')
plt.hlines(0,0,len(lr.coef_))
plt.ylim(-25,25)
plt.legend(loc=1)
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso().fit(x_train,y_train)
print('{:.2f}, {:.2f}'.format(lasso.score(x_train,y_train),lasso.score(x_test,y_test)))
```

```python
lasso001 = Lasso(alpha=.01,max_iter=100000).fit(x_train, y_train)
print('{:.2f}, {:.2f}'.format(lasso001.score(x_train,y_train),lasso001.score(x_test,y_test)))
```

```python
lasso0001 = Lasso(alpha=.0001, max_iter=100000).fit(x_train,y_train)
print('{:.2f}, {:.2f}'.format(lasso0001.score(x_train,y_train),lasso0001.score(x_test,y_test)))
print(np.sum(lasso0001.coef_ != 0))#CARACTERISTICAS USADAS
```

```python
plt.plot(lasso.coef_, 's', label='lasso alpha 1')
plt.plot(lasso001.coef_, '^', label='lasso alpha 0.01')
plt.plot(lasso0001.coef_, 'v', label='lasso alpha 0.0001')

plt.plot(ridge01.coef_, '.', label='ridge alpha 0.01')
plt.legend(ncol=2, loc=(0,1.05))
# ncol=numeros de columnas///loc=()=poicion de la leyenda 
plt.ylim(-25,25)
plt.xlabel('coefficient index')
plt.ylabel('coefficient magnitud')
```
