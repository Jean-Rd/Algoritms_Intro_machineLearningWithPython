```python
import mglearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
_X, _Y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1,3, figsize=(10,3))

for i, ax in zip([1,3,9], axes):
    
    clf = KNeighborsClassifier(n_neighbors=i).fit(_X,_Y)
    #mglearn.plots.plot_2d_separator(clf)
    mglearn.plots.plot_2d_separator(clf, _X, fill=True, ax=ax, eps=.5, alpha=.4)
    mglearn.discrete_scatter(_X[:,0], _X[:,1], _Y, ax=ax)
    ax.set_title('{} neighbor(s)'.format(i))
    ax.set_xlabel('feature 0')
    ax.set_ylabel('feature 1')
    
axes[0].legend(loc=3)
plt.show()
```
