```
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```
func = lambda th: -np.sin(((th[0]**2)/2)-(0.25*(th[1]**2))+3)*np.cos(2*th[0]+1-(np.e**th[1]))
```

```
_X = np.linspace(-2,2,100)
_Y = np.linspace(-2,2,100)
_Z = np.zeros((100,100))
```

```
for ix, x in enumerate(_X):
  for iy, y in enumerate(_Y):
    _Z[iy,ix] = func([x,y])

plt.contourf(_Y, _X, _Z, 100)

lr = .1
h = .001
grad = np.zeros(2)
t = np.random.rand(2) * 4 - 2

for _ in range(5000):

  for it in range(len(t)):

    tt = np.copy(t)
    tt[it] = tt[it] + h
    NDER = (func(tt) - func(t)) / h
    grad[it] = NDER

  t = t - (lr * grad)

  if _ % 10 == 0:

    plt.plot(t[0], t[1], '.', c='white')

plt.plot(t[0], t[1], 'o', c='black')
```
