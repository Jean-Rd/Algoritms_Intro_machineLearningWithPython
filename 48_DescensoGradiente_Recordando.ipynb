```
# Descenso del gradiente

import numpy as np
import matplotlib.pyplot as plt

func = lambda th: (th[0]**2 + 3*th[1]**2) * (np.e**(1 - th[0]**2 - th[1]**2)) * -np.cos(np.e*(th[1] + th[0]))

_X = np.linspace(-2, 2, 100)
_Y = np.linspace(-2, 2, 100)
_Z = np.zeros((100,100))
```

```
for ix, x in enumerate(_X):

  for iy, y in enumerate(_Y):

    _Z[iy, ix] = func([x,y])

plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()

lr = .01
delta_X = 0.001
grad = np.zeros(2)
t = np.random.rand(2) * 4 - 2

for _ in range(1000):

  for it in range(len(t)):

    t_copy = np.copy(t)

    t_copy[it] = t_copy[it] + delta_X

    NDER = (func(t_copy) - func(t)) / delta_X

    grad[it] = NDER

  t = t - (lr * grad)

  if _ % 15 == 0:

    plt.plot(t[0], t[1], '.', c='white')

plt.plot(t[0], t[1], 'X', c='black')
plt.show()
```
