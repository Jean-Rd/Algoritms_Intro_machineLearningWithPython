```
import seaborn as sns
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()

print(iris.DESCR)
```

```
sns.distplot(iris.data[:,0], kde=True)
```

```
sns.distplot(iris.data[:,0], kde=False)
```

```
sns.distplot(iris.data[:,2], kde=False)
```

```
sns.kdeplot(iris.data[:,0], shade=True)
```

```
sns.kdeplot(iris.data[:,3], shade=True)
```

```
sns.kdeplot(iris.data[:,0], iris.data[:,3])
```

```
sns.kdeplot(iris.data[:,2], iris.data[:,3], shade=True, cbar=True)
```

```
sns.jointplot(x=iris.data[:,2], y=iris.data[:,3], kind='kde', shade=True)
```

```
sns.jointplot(iris.data[:,0], iris.data[:,3], shade=True, kind='kde')
```

```
sns.distplot(iris.data[:,0], label='Length', kde=False)
sns.distplot(iris.data[:,1], label='Wifth', kde=False)
sns.distplot(iris.data[:,2], label='Length 2', kde=False)
sns.distplot(iris.data[:,3], label='width 2', kde=False)
plt.legend(loc='best')
```

```
sns.kdeplot(iris.data[:,0], label='length', shade=True)
sns.kdeplot(iris.data[:,1], label='width', shade=True)
sns.kdeplot(iris.data[:,2], label='length 2', shade=True)
sns.kdeplot(iris.data[:,3], label='width 2', shade=True)

plt.legend(loc='best')
```
