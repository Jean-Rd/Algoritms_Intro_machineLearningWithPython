```
from sklearn.datasets import load_breast_cancer
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
import pandas as pd
```

```
breast_cancer = load_breast_cancer()
```

```
X = breast_cancer.data
Y = breast_cancer.target

Y = np.reshape(Y, (569, 1))

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(_X, Y, random_state=1)

axis_x = range(x_test.shape[0])
axis_y = y_test

```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error

rgl = np.linspace(.01, 1, 10)

_mse = []

for _ in rgl:

  logreg = LogisticRegression(C = _)
  logreg.fit(x_train, y_train)
  pred = logreg.predict(x_test)

  _mse.append(mean_squared_error(y_test, pred))
```

```
df_mse = pd.DataFrame({'MSE LogReg':_mse, 'C':rgl})
df_mse.head()

_C = [rgl[1], rgl[3]]

df_mse[df_mse['MSE LogReg'] == df_mse['MSE LogReg'].min()]

```

```
LR = LogisticRegression(C = _C[0]).fit(x_train, y_train)
predl = LR.predict(x_test)

fig, ax = plt.subplots()

ax.plot(axis_x, axis_y, '-', label='Data')
ax.plot(axis_x, predl, 'o', label='Prediction')
fig.set_size_inches(15,8)
plt.show()
```
