```python
import pandas as pd
df_cancer = pd.DataFrame(cancer.data,columns=cancer.feature_names)
df_cancer.head()
```

```python
plt.plot(cancer.data[:,0],cancer.data[:,1],'x',alpha=.8)
plt.show()
```

```python
_X = cancer.data[:,0]
_Y = cancer.data[:,1]
_Z = np.zeros((569,569))

func = lambda th: (np.sin(th[0]*2)*np.cos(th[1]*th[0]) + np.e)
```

```python
for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy,ix] = func([x,y])
```

```python
plt.contourf(_X,_Y,_Z)
plt.colorbar()
plt.show()
```
