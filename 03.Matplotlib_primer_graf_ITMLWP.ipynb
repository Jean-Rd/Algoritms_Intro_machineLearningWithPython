```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# creamos nuestros datos para el eje x
x = np.linspace(-10,10,100)
# para hallar nuestra funcion de onda seno aplicamos esta funcion trigonometrica a esta variable 'x'
y = np.sin(x)
plt.plot(x, y, marker='x')
```
