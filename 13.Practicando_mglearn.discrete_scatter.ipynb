```python
'''
usaremos mglearn.make_forge(x,y,target)
para clasificar y ver la distribucion de nuestros datos
'''
import mglearn
import matplotlib.pyplot as plt

x,y = mglearn.datasets.make_forge()

mglearn.discrete_scatter(x[:,0],x[:,1],y)
plt.xlabel('First feature')
plt.ylabel('Second feature')
plt.legend(['Class 0', 'Class 1'], loc=3)
plt.show()
#clasifica las classes de datos por colores 
```

```python
import pandas as pd
import numpy as np

pokemon = pd.read_csv('csv/Pokemon.csv')
pokemon.head()
```

```python
np.array(pokemon.iloc[:,6])
np.array(pokemon.iloc[:,7])
mglearn.discrete_scatter(np.array(pokemon.iloc[:,6]),np.array(pokemon.iloc[:,7]),np.array(pokemon.iloc[:,2]))
plt.legend(set(np.array(pokemon.iloc[:,2])), loc=1)

#aca vemos como se distribuyen grandes cantiudades de datos y que las figuras cambian aletoricamente
```
