```python
from sklearn.model_selection import train_test_split

_X, _Y = mglearn.datasets.make_forge()
 
_xtrain, _xtest, _ytrain, _ytest = train_test_split(_X,_Y,random_state=0)
_xtrain
# con random_state=0, nuestros valores no camvian aletoricamente
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn.fit(_xtrain, _ytrain)

knn.predict(_xtest)
```

```python
knn.score(_xtest,_ytest)
```
