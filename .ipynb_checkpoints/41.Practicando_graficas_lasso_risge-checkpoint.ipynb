```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
iris = load_iris()
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=0)

knn = KNeighborsClassifier(n_neighbors=3).fit(x_train, y_train)
linear_svm = LinearSVC(C=.1).fit(x_train, y_train)
```

```python
print(knn.score(x_test, y_test))
print(linear_svm.score(x_test, y_test))
```

```python
train_ = []
test_ = []
train_1 = []
test_1 = []

for i in range(1,11):
    
    clf = KNeighborsClassifier(n_neighbors=i).fit(x_train, y_train)
    clf2 = LinearSVC(C=i).fit(x_train, y_train)
    
    train_.append(clf.score(x_train, y_train))
    test_.append(clf.score(x_test, y_test))
    train_1.append(clf2.score(x_train, y_train))
    test_1.append(clf2.score(x_test, y_test))
    
plt.plot(range(1,11), train_, c='blue', marker='.', label='train knn')
plt.plot(range(1,11), test_, c='cyan', marker='*', label='test knn')
plt.plot(range(1,11), train_1, c='red', marker='.',label='train linearSVC')
plt.plot(range(1,11), test_1, c='green', marker='*', label='test linearSVC')
plt.legend(loc=(1.01,0.3))
plt.grid()
plt.show()
```

```python
_train=[]
_test=[]
_train1=[]
_test1=[]
settings= np.linspace(0,1,10)

for i in setting:
    
    ridge = Ridge(alpha=i).fit(x_train, y_train)
    lasso = Lasso(alpha=i).fit(x_train, y_train)
    
    _train.append(ridge.score(x_train, y_train))
    _test.append(ridge.score(x_test, y_test))
    _train1.append(lasso.score(x_train, y_train))
    _test1.append(lasso.score(x_test, y_test))
    
plt.plot(settings, _train, marker='.', c='blue', label='train ridge')
plt.plot(settings, _test, marker='*', c='cyan', label='test ridge')
plt.plot(settings, _train1, marker='.', c='red', label='train lasso')
plt.plot(settings, _test1, marker='*', c='green', label='test lasso')
plt.grid()
plt.legend(loc=(1.1,.3))
plt.show()
```
