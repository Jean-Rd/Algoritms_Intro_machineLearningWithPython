```python
import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np
from sklearn.datasets import load_iris
import mglearn
import pandas as pd
```

```python
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=1)
'''
como ya sabemos debemos llamar la libreria sklearn.model_selection e importar train_test_split
una ves declaramos las variables de estas en la ultima parte debemos agregar random_state(estado aletorico)
igual a cero para que nuestros datos no canvien de poscion 
x_train--->contiene el 75% de los datos de iris.data
x_test-->contiene el 25% de los datos de iris.data
y_train-->contiene el 75% de los datos de target(objeto'son los tipos de rosas')
y_test-->contiene el 25% de los datos de target
todas estas variables estan distribuidas aletoricamente.
'''
```

```python
'''
importaremos un modulo que se uso para london, donde se predice el costo de los departamentos
'''
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
#ahora pasamos los datos a knn.fit que es un metodo de knn; este metodo fit relaciona datos nuevos
#ingreasdos y los ubica sgun el orden de njuestro x_train, y lo casifica como una especie se debe pasar
#una lista de 2d con los datos completos y este lo predice en una reg lineal en el fondo
knn.fit(x_train, y_train)
```

```python
#creamos nuestro nuevo dato para predecir su tipo con length and width of sepal and petal
x_new = np.array([[5,2.9,1,0.2]])
print('x_new shape:\n{}'.format(x_new.shape))
```

```python
#ahora predecimos con el metodo knn.predict
prediction = knn.predict(x_new)
print('prediction: {}'.format(prediction))
print('prediction target name: {}'.format(iris.target_names[prediction]))
# y asi nos devuelve que nuestros datos ingresados por teclado es de tipo 0, por lo tanto es setosa
```
