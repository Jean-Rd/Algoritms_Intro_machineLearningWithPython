```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
func = lambda th: -np.sin(((th[0]**2)/2)-(0.25*(th[1]**2))+3)*np.cos(2*th[0]+1-(np.e**th[1]))
'''
cambiamos la a x ^ y por th(0), th(1) respectivamente.
th es un array de shape [,2], una  fila dos columnas es lo que debemos psar como parametros
'''
```

```python
_X = np.linspace(-2,2,100)
_Y = np.linspace(-2,2,100)
_Z = np.zeros((100,100))
'''
creamos los ejes x ^ y, cada una con 100 elementos en un rango de -2,2
en z creamos nuestra matriz irentidad, donde guardaremos nuestros datos
'''
```

```python
for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = func([x, y])
        
p = np.array([5,3])
print(func([5,3]))
print(func(p))
```

```python
'''
estas dos funciones son equivalentes, solo que con enumerate nos ahoramos los indices 
este nos lo da en una variable

for ix in range(len(_X)):
    for iy in range(len(_Y)):
        _Z[iy,ix] = func([_X[iy],_Y[ix]])

func([5,3])'''
```

```python
'''
graficamos nuestros datos tridimensionales, ahora solo vemos la densidad de nuestro datos
esta vista es desde arriba
'''
plt.contour(_X,_Y,_Z)
plt.show()
```

```python
#ahora aumentamos los parametros a contour para que grafique mas lineas
plt.contour(_X,_Y,_Z,100)
```

```python
plt.contourf(_X,_Y,_Z)
plt.colorbar()
#Superficie trimensional solida

#generamos un punto en nuestra grafica; blanco son los puntos altos y guindo son los bajos

_T = np.random.rand(2) *4 -2
#multiplicamos por 4 y resstamos 2, por que numpy.random solo nos da valors de 0-1 
_T

plt.plot(_T[0],_T[1],'X',c='black')

_TM = np.copy(_T) # copiamos nuestros datos por que alteraremos en el gradient desenct
_D = 0.0001 # diferenciacion value, deltaX.
_RL = 0.015 # ratio de aprendizaje
_GRAD = np.zeros(2) # donde guardamos nuestra pendiente gradiente

for _ in range(10000):
    for it, th in enumerate(_T):
        
        _TM = np.copy(_T)
        _TM[it] = _TM[it] + _D # v _TM[it] = th +_D
        _DINF = (func(_TM)-func(_T))/_D
        _GRAD[it] = _DINF
        #print(_TM[it],th) iguales
        
    _T = _T - _RL*_GRAD
    
    if _ % 10 == 0:
        
        plt.plot(_T[0],_T[1],'.',c='yellow')

plt.plot(_T[0],_T[1],'o',c='white')
plt.show()
```
