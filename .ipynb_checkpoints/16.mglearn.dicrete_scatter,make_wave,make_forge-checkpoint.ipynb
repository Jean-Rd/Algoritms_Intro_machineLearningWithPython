```python
import mglearn 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = mglearn.datasets.make_forge()
```

```python

# DISPERSION DE DATOS.PLOT

mglearn.discrete_scatter(x[:,0],x[:,1],y)
plt.legend(['Class 0', 'Class 1'], loc=4)
plt.xlabel('First feature')
plt.ylabel('Second feature')
print('x.shape: {}'.format(x.shape))
'''
en mglearn.discrete_scatter(x,y,target)
primer parametro es datos para axis x
second parametro is data fro axis y
three parametro is target of the dates
en el tercero ponemos el tipo de datos con que se clasifican por ejemplo
en load iris, si ingresamos x-length sepal and y-width sepal in 3parametre seria target
donde seleccionaria por color a cada tipo de iris
'''
```

```python
x2, y2 = mglearn.datasets.make_wave(n_samples=40)
plt.plot(x2,y2,'o')
plt.ylim(-3,3)
plt.xlim(-3,3)
plt.xlabel('feature')
plt.ylabel('target')
```

```python
import pandas as pd
from sklearn.datasets import load_iris
df = pd.DataFrame(x,y)
df

iris = load_iris()

mglearn.discrete_scatter(iris.data[:,0],iris.data[:,1],iris.target, alpha=.8)
plt.legend(iris.target_names, loc=1)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```

```python
from sklearn import linear_model
# MINIMOS CUADRADOS ORDINARIO REF
reg = linear_model.Ridge(alpha=.5)
reg.fit([[0,0],[0,0],[1,1]],[0,.1,1])
reg.coef_
#coeficientesw de la regresion
reg.intercept_

# REGRESION LINEAL NORMAL 

reg2 = linear_model.LinearRegression()
reg2.fit([[0,0],[1,1],[2,2]],[0,1,2])
reg2.coef_
```
