```python
# decenso del gradiente

import numpy as np
import matplotlib.pyplot as plt
```

```python
func = lambda th: -np.sin(((th[0]**2)/2)-((th[1]**2)/4)+3)*np.cos(2*th[0]+1-np.e**th[1])

func([5,3])
```

```python
_X = np.linspace(-2,2,100)
_Y = np.linspace(-2,2,100)
_Z = np.zeros((100,100))


for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = func([x,y])


plt.contourf(_X,_Y,_Z,100)
plt.colorbar()

theta = np.random.rand(2) * 4 - 2
grad = np.zeros(2)

plt.plot(theta[0], theta[1], 'X', c='red')

_T = np.copy(theta)
_H = 0.00001
_RL = 0.01

for _ in range(1000):
    
    for ith, th in enumerate(theta):
        
        _T = np.copy(theta)
        
        _T[ith] = _T[ith] + _H
        
        _der = (func(_T) - func(theta))/_H
        
        grad[ith] = _der
        
        #print(theta)
        
    theta = theta - _RL * grad
        
        #print(theta)
        
   #print(func(theta))
    
    if _ % 100 == 0:
        
        plt.plot(theta[0], theta[1], '.', c='black')
        

plt.plot(theta[0], theta[1], 'x', c='black')        
plt.show()
```
