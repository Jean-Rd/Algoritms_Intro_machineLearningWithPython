```
from sklearn.datasets import load_boston
import numpy as np
import matplotlib.pyplot as plt
```

```
boston = load_boston()
print(boston.DESCR)
```

```
_X = np.array(boston.data[:,5])
_Y = np.array(boston.target)

plt.scatter(_X, _Y, alpha=.5)

_X = np.array([np.ones(len(_X)), _X]).T

b = np.linalg.inv(_X.T @ _X) @ _X.T @ _Y

plt.plot([4,9], [b[0] + b[1] * 4, b[0] + b[1] * 9], c='purple')

# IMPORTANTE HACER SOLO EN UNA CASILLA SI USAMOS JUPYTER O COLAB
```
